**2.0** Fit a linear model predicting the number of views (views), from the length of a video (length) and its average user rating (rate).

```r
model = lm(log(views) ~ log10(length) + rate, data= Data)
```

**3.0** Using diagnostic plots, background knowledge, and statistical tests, assess all 6 assumptions of the CLM.  When an assumption is violated, state what response you will take.

CLM Assumption 1 : **Linear in parameters**

We assume that the population model for number of views is a linear in parameters with the covariates length of the video and rating of the video. Our model construction above does satify this assumption.

CLM Assumption 2: **Random Sampling**

We assume that the data collected is ranndomly sampled from the population and that the error terms are i.i.d. We cannot observe the error terms in the model. However, given the diversity of videos on online platforms and the large number of observations (~1000) in the data set we can assume that the sampling is random.

CLM Assumption 3: **Sample variation and no Perfect collinearity**

We must have some variation in the sample data and there should not be a perfect correlation between the covariates. We see from the EDA of the covariates above that there is variation in the data used for the analysis. We check the perfect collinearity assumption below. 

```r
cor(log10(Data$length), Data$rate, use="complete.obs")
plot(log10(Data$length), Data$rate, col = "blue", main="Figure 1: Ratings vs length of videos")
```

The correlation between the covariates is significantly less than one. Hence we do not have perfect collinearity between the covariates. This is also evident from figure 1 (ratings vs length of videos)

```r
plot(model, which = 1, col="blue", lwd=2, main = "Figure 2")
mean(model$residuals)
```

CLM Assumption 4: **Zero conditional mean**

The mean of the residuals conditional on the values of dependant variables is zero. i.e. there is no functional relationship between X and u 
$$E(u|x) = 0$$

The residulas vs fitted diagonstics plot (Figure 2) shows a mean fit line that is fairly close to zero and is approximately flat through most of the domain of fitted values. This shows that the zero conditional mean assumption is not significantly violated.

CLM Assumptionn 5: **Homoskedasticity**
