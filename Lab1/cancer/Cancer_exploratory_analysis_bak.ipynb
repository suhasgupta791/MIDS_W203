Looks like from above data, the medianAge is the mean of MedianAgeMale & MedianAgeFemale when the medianAge is a valid entry. The 30 entries in the second filter do not seem to be valid for Medianage but may be valid data based on MedianAgeMale & MedianAgeFemale.

```r
MedianAgeCorrected = (cancerData$MedianAgeMale + cancerData$MedianAgeFemale)/2
scatterplotMatrix(~ MedianAgeMale + MedianAgeFemale + MedianAgeCorrected, data = cancerData)
boxplot(MedianAgeCorrected)
```

The data in AvgHouseholdSize variables contains entries less than 1 that doesn't make sense (the average household size in a county must be greater than 1) 
It may be necessary to remove entries from the analysis that have AvgHouseholdSize < 1

```r
summary(cancerData$AvgHouseholdSize)
hist(cancerData$AvgHouseholdSize)
head(cancerData[cancerData$AvgHouseholdSize<1,])
```

```r
scatterplotMatrix(~ log(avgAnnCount,10) + povertyPercent + log(popEst2015,10) + PctPrivateCoverage 
                  + PctPublicCoverage + MedianAgeCorrected + AvgHouseholdSize + deathRate, data = cancerData )
```
